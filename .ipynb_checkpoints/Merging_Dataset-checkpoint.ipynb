```python
import pandas as pd

data2 = pd.read_excel('CustomerBehaviorData.xlsx')
data3 = pd.read_excel('CustomerDemographics.xlsx')

# Select the columns you want to merge from data3
data3_merged = data3[["ID", "account_type", "gender", "age", "region_code"]]

# Merge data2 with the selected columns from data3
result = pd.merge(data2, data3_merged, on='ID', how='left', indicator=True)

result.to_excel("Results.xlsx", index = False) 
```

```python
data1 = pd.read_excel('CreditConsumptionData.xlsx')

result = pd.read_excel('Results.xlsx')

result = result.drop('_merge', axis=1)

dataset = pd.merge(data1, result, on='ID', how='inner', indicator=True)
```
