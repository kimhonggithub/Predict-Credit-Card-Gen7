```python
#import
import pandas as pd
import matplotlib.pyplot as mt
```

```python
#read data
df = pd.read_excel('dataset.xlsx')
#check data length
len(df)
```

```python
#print data head
df.head()
```

```python
# Check for null values
df.isnull().sum()
```

```python
# drop rows with missing values
df = df.dropna(subset=['cc_cons_may', 'dc_cons_jun', 'cc_count_apr', 'personal_loan_closed', 'investment_3', 'debit_count_apr', 'loan_enq', 'emi_active', 'account_type', 'gender', 'region_code'])
```

```python
# Check for null values
df.isnull().sum()
```

```python
# check percentage of missing values
df.duplicated().sum()
```

```python
# check data types of columns
df.dtypes
```

```python
# convert data types to the correct format
convert_dict = {
    'cc_count_apr': int,
    'personal_loan_closed': int,
    'debit_count_apr': int,
}
df = df.astype(convert_dict)
```

```python
# check data types of columns
df.dtypes
```

```python
# check for negative values (outliers)
df._get_numeric_data().lt(0).sum()
```

```python
# set negative values to 0
for x in df.index:
    if df.loc[x, 'investment_4'] < 0:
        df.loc[x, 'investment_4'] = 0
```

```python
# check for negative values (outliers)
df._get_numeric_data().lt(0).sum()
```

```python
neworder = [
    'ID','account_type', 'gender', 'age', 'region_code', 'cc_cons_apr',
             'dc_cons_apr', 'cc_cons_may', 'dc_cons_may', 'cc_cons_jun',
             'dc_cons_jun', 'cc_count_apr', 'cc_count_may', 'cc_count_jun',
             'dc_count_apr', 'dc_count_may', 'dc_count_jun', 'card_lim',
             'personal_loan_active', 'vehicle_loan_active', 'personal_loan_closed',
             'vehicle_loan_closed', 'investment_1', 'investment_2', 'investment_3',
             'investment_4', 'debit_amount_apr', 'credit_amount_apr',
             'debit_count_apr', 'credit_count_apr', 'max_credit_amount_apr',
             'debit_amount_may', 'credit_amount_may', 'credit_count_may',
             'debit_count_may', 'max_credit_amount_may', 'debit_amount_jun',
             'credit_amount_jun', 'credit_count_jun', 'debit_count_jun',
             'max_credit_amount_jun', 'loan_enq', 'emi_active','cc_cons'
]
df = df[neworder]
```

```python
#split dataset into train and test
#get data with not null cc_cons as train data
train_df = df[df['cc_cons'].notnull()]
#get data with null cc_cons as test data
test_df = df[df['cc_cons'].isnull()]
```
