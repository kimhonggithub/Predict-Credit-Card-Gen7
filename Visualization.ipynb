```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

```python
df = pd.read_excel('datavisualize.xlsx')
```

```python

bins = [25, 35, 45, 55, 100]
# Custom age group labels
age_group_labels = ['25-34', '35-44', '45-54', '55 and above']

gender_age_groups = df.groupby(['gender',pd.cut(df['age'],bins),'account_type']).size().unstack(fill_value=0)

men_data = gender_age_groups.loc['M']
women_data = gender_age_groups.loc['F']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6),sharey=True)

# plot data for Male
men_data.plot(kind='bar', width = 0.35, ax = ax1)
ax1.set_xlabel('Age group')
ax1.set_ylabel('Count')
ax1.set_title('Gender = M')
ax1.set_xticklabels(age_group_labels, rotation=0)
ax1.get_legend().remove()
# ax1.legend(title='Card Type', loc='upper right')

# Plot for women
women_data.plot(kind='bar', width=0.35, ax=ax2)
ax2.set_xlabel('Age Groups')
ax2.set_ylabel('Count')
ax2.set_title('Gender = F')
ax2.get_legend().remove()
ax2.set_xticklabels(age_group_labels, rotation=0)
# ax2.legend(title='Card Type', loc='upper right')

plt.legend(title='Account Type', loc='center right',bbox_to_anchor=(1.3,0.5), fancybox=True, shadow=True)
fig.suptitle('Age Group of Men and Women Using Credit and Debit Cards')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
sum_of_count_apr = pd.DataFrame()
sum_of_count_apr['credit'] = [df['cc_count_apr'].sum()]
sum_of_count_apr['debit'] = [df['dc_count_apr'].sum()]

sum_of_count_may = pd.DataFrame()
sum_of_count_may['credit'] = [df['cc_count_may'].sum()]
sum_of_count_may['debit'] = [df['dc_count_may'].sum()]
sum_of_count_may

sum_of_count_june = pd.DataFrame()
sum_of_count_june['credit'] = [df['cc_count_jun'].sum()]
sum_of_count_june['debit'] = [df['dc_count_jun'].sum()]
sum_of_count_may
```

```python
sns.set_style('whitegrid')
colors = sns.color_palette('pastel')

# Combine the DataFrames into a single DataFrame for plotting
summary_data = pd.concat([sum_of_count_apr, sum_of_count_may, sum_of_count_june])

# Plot the bar graph
# plt.figure().set_figwidth(15)
# plt.figure().set_figheight(15)
ax = summary_data.plot(kind='bar', color=colors, rot=0)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Transaction Count')
plt.title('Credit and Debit Card Transaction Count for Each Month')

# Add legend
plt.legend(title='Card Type', labels=['Credit', 'Debit'], loc='upper right',bbox_to_anchor=(1.3,0.5))
# Set x-axis labels to April, May, and June
plt.xticks(range(len(summary_data.index)), ['April', 'May', 'June'])


# Show the plot
plt.show()
summary_data
```

```python
# df['region_code'].unique()
occurrences = df['region_code'].value_counts().head(10)

total_count = occurrences.sum()

# total_count
percentages = occurrences / total_count * 100
percentages_sorted = percentages.sort_values(ascending=False)
# percentages_sorted

top_10 = percentages_sorted
# top_10

top10_result_df = pd.DataFrame(top_10)
top10_result_df.rename(columns={'count': 'percentage'}, inplace=True)
top10_result_df.reset_index(drop=False, inplace=True)
top10_result_df

```

```python
colors = sns.color_palette("pastel")
plt.figure(figsize=(5, 5))
plt.pie(top10_result_df['percentage'], labels=top10_result_df['region_code'], autopct='%1.1f%%', startangle=140, colors=colors)
plt.axis('equal')
plt.title('Top 10 region and its percentage')
plt.show()
```
